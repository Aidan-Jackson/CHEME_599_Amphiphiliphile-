First, we will import the test image to be used in the workflow.

```python
import numpy as np
import matplotlib.pyplot as plt

test_image = plt.imread('project_image.jpg') #Reads in an example test image
```

The first goal in the analysis of these images is location and counting of the blobs in a single image. Later it is desired to track blobs through a series of images that are taken at multiple times from a single physical sample.

The main blob_detection_method.py gives the x and y coordinates, size, and total count of the white and black blobs as well as lenses. 

```python
import amph
from amph import blob_detection_method

min_thr = 0.5 #sets parameters of of blob_dog used in blob_detection_method.blobs
sig_max = 15

x, y, r, blob_count, xl, yl, rl, lens_count, xd, yd, rd, numd, markers = blob_detection_method.blobs(test_image,sig_max,min_thr)
```

The *blobs* function directly determines the number of white blobs and lenses. This are explicitly reported below.

```python
print('There were ' + str(blob_count) + ' light blobs, ' + str(numd) + ' dark blobs, & ' +str(lens_count)+ ' lenses found in the image.')
```

We can also take the found particles and overlay them on top of the original image. 

```python
blob_detection_method.graph_blobs(test_image, x, y, r, xl, yl, rl, xd, yd, rd)


fig, ax = plt.subplots(figsize=(5,5))

ax.imshow(markers, cmap='gray')
ax.axis('off')
```
